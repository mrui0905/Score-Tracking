```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
def create_plot(mode, game, type):

    path = 'data/' + game+ '/' + mode + '_' + type + '.csv'
    
    data = pd.read_csv(path)
    data.columns = ['Date', 'Time', 'Score'] if type == 'raw' else ['Date', 'Score']

    data['Date'] = pd.to_datetime(data['Date'], format = '%m/%d/%y')

    date = np.asarray(data['Date'])
    score = np.array(data['Score'])

    plt.scatter(date, score)
    plt.title((mode+' '+type).title())
    plt.figure(figsize=(50,12))
    plt.show()
```

```python
# Zetamac Graphs
game = 'zetamac'
for mode in ['addition', 'subtraction', 'multiplication', 'division', 'combined']:
    for type in ['raw', 'avg']:
        create_plot(mode, game, type)
```
